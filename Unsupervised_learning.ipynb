**For the given Data, apply Niave Bayes classification algorithm and Improve accuracy using feature selection and parameter tuning**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
#load and read file
df = pd.read_csv('/content/heart.csv');
```

```python
df
```

```python
df.head()
```

Lets divide columns into two types -- dependent(or target variable) and independent variable(or feature variables).

```python
x= df.drop('target',axis=1)
y = df['target']

print('x=',x)
print('y=',y)
```

```python
#split into train and test set
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)

print('xtrain=',x_train)

print('ytrain=', y_train)

print('xtest=',x_test)

print('ytest=',y_test)
```

Multinomail NB

```python
model = MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
#make predictions
y_pred = model.predict(x_test)
```

```python
print('actual labels = ')
print(y_test)

print('predict labels = ')
print(y_pred)
```

```python
#evaluating model
accuracy = accuracy_score(y_test,y_pred)*100
print('Accuracy',accuracy)
```

Gaussian NB

```python
from sklearn.naive_bayes import GaussianNB
model_Gas = GaussianNB()
model_Gas.fit(x_train,y_train)
y_pred1 = model_Gas.predict(x_test)

#make predictions
#y_pred1 = model_Gas.predict(x_test)
```

```python
print('actual labels = ')
print(y_test)

print('predict labels = ')
print(y_pred1)
```

```python
#evaluating model
accuracy1 = accuracy_score(y_test,y_pred1)*100
print('Accuracy',accuracy1)
```

Bernoulli NB

```python
from sklearn.naive_bayes import BernoulliNB

model_B = BernoulliNB(fit_prior=False)
model_B.fit(x_train,y_train)
y_pred2 = model_B.predict(x_test)
```

```python
print('actual labels = ')
print(y_test)

print('predict labels = ')
print(y_pred2)
```

```python
#evaluating model
accuracy2 = accuracy_score(y_test,y_pred2)*100
print('Accuracy',accuracy2)
```
